```python
import numpy as np

# Set random seed for reproducibility
np.random.seed(1234)

# Generate Y values randomly from a normal distribution
Y = np.random.normal(loc=0, scale=1, size=1000)

# Calculate corresponding X values based on the line x = 2y
X = 0.5 * Y + np.random.normal(loc=0, scale=0.5, size=1000)

# Combine X and Y into a single array
data_1 = np.vstack((X, Y)).T



# Generate Y values randomly from a normal distribution
Y = np.random.normal(loc=0, scale=1, size=1000)

# Calculate corresponding X values based on the line x = 2y
X = 2 * Y + np.random.normal(loc=0, scale=0.5, size=1000)

# Combine X and Y into a single array
data_2 = np.vstack((X, Y)).T
```

```python
plt.scatter(data_1[:,0], data_1[:,1])
plt.scatter(data_2[:,0], data_2[:,1])
```

```python
def transform(data, transfrom_data):
    cov_mat = np.dot(data.T, data)/data.shape[0]
    
    eig_values, eig_vectors = np.linalg.eigh(cov_mat)
    
    return np.dot(eig_vectors.T, transfrom_data.T).T
```

```python
data_1_transformed = transform(data_1, data_1)

data_2_transformed = transform(data_2, data_2)
```
